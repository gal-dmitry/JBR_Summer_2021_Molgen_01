**Particle swarm optimization**

- каждый агент хранит свое лучшее решение
- лучшее решение среди всех агентов доступно каждому
- каждый из агентов конструирует новую молекулу на основе комбинации личного и коллективного лучшего решения
- произведенные молекулы оценивается с помощью fitness-функции
- лучшие решения обновляются 

Таблица с примерами алгоритмов из статьи [The quest for novel chemical matter and the contribution of computer-aided de novo design, Bernard Pirard](https://pubmed.ncbi.nlm.nih.gov/22647201/)

### RNN

Статья: [SMILES Enumeration as Data Augmentation for Neural Network Modeling of Molecules”
Bjerrum, 2017](https://arxiv.org/pdf/1703.07076.pdf)


**1. Архитектура:**

В данной статье по набору различных smiles-строк одной и той же молекулы предсказывались ее свойства. Строка smiles подавалась как последовательность на вход рекуррентной нейронной сети - **RNN**. Далее, выход из RNN-сети подавался на вход полносвязной нейронной сети, которая предсказывала значение log $IC_{50}$.


![](pictures/1.png)

Рекуррентная нейронная сеть состоит из **LSTM-ячеек** - ячеек “долгой краткосрочной памяти”. Каждая такая ячейка имеет 3 фильтра:
- фильтр забывания $f_t$
- входной фильтр $i_t$
- выходной фильтр $o_t$

Каждый из фильтров принимает на вход новый элемент последовательности $x_t$, а также выход из предыдущей ячейки $h_{t-1}$. Кроме этого, LSTM-ячейка принимает на вход состояние $C_{t-1}$ с предыдущей ячейки. Фильтр забывания решает, какая часть информации будет забыта. Входной фильтр решает, какую часть информации следует обновить. Итоговое значение $h_t$ получается после применения выходного фильтра.

![](pictures/03.png)

![](pictures/3.png)

**2. Особенности подхода в данной статье:**

Строковое представление молекулы - SMILES - может быть не единственным.
Поэтому в процессе предобработки данных smiles приводят к каноническому виду или оставляют только smiles, который встречается чаще остальных.


![](pictures/1_2.png)

В данной статье же наоборот, авторы генерируют различные smiles для одной молекулы, чтобы использовать это в качестве техники аугментации данных (как  поворот картинок для сверточных нейросетей). В статье было показано, что такой подход приводит к улучшению результатов.

Сравнивались 2 модели **Canonical Model** и **Enumerated Model**, натренированные на одном и том же датасете с аугментацией и без аугментации соответственно. Для каждой нейросети производилась оптимизация гиперпараметров, поэтому итоговая архитектура имеет следующий вид:

### VAE


Статья: [Automatic Chemical Design Using a Data-Driven Continuous Representation of Molecules, Gómez-Bombarelli, 2016](https://arxiv.org/pdf/1610.02415.pdf)

**1. Архитектура:**

Вариационный автоэнкодер состоит из двух частей: 
- Encoder переводит объект (в нашем случае smiles-строку) в латентное континуальное пространство. Encoder учится делать это с ограничениями на разницу распределений (KL-дивергенция). 
- Decoder пытается восстановить объект из латентного пространства.

![](pictures/04.png)

**2. Особенности подхода в данной статье:**

- Задача: генерировать молекулы с желаемыми свойствами
- Для предсказания свойств добавлена нейросеть-предиктор. Она предсказывает свойства молекулы на основе ее представления в латентном пространстве. Ошибка предсказания свойств молекулы включена в функцию потерь.
- Невалидные смайлс строки удаляются с помощью RDKit на выходе

![](pictures/6.png)

### GAN

Cтатья: [Objective-Reinforced Generative Adversarial Networks (ORGAN) for Sequence Generation Models, Guimaraes, 2017](https://arxiv.org/pdf/1705.10843.pdf)

**1. Архитектура GAN (левая часть рисунка):**

![](pictures/7.png)

$G_θ$ - генератор, который учится выдавать последовательность $Y_{1:T} = (y_1, ..., y_T)$ длины T (в данном случае smiles-строку), так чтобы дискриминатор принял ее за настоящую.

$D_φ$ - дискриминатор, который учится отличать реальные последовательности от сгенерированных.

Модели обучаются с чередованием, оптимизируя следующую функцию:


![](pictures/8.png)

**2. Особенности подхода в данной статье (правая часть рисунка):**


В данном случае (так как для дискретных данных генерация последовательностей не может быть дифференцируема) $G_θ$ обучается как агент в задаче RL, используя алгоритм **REINFORCE**. Получая строку-последовательность длины t - $Y_{1:t}$, которую можно рассматривать как состояние $s_t$, агент должен возвращать следующий символ последовательности - $y_{t+1}$, пытаясь максимизировать суммарную награду $R$, полученную за итоговую последовательность $Y_{1:T}$:

![](pictures/9.png)

Для этого используется алгоритм **N-time Monte Carlo search**: на шаге $t$ с помощью $G_θ$ сэмплируется недостающая часть последовательности - $Y_{t+1:T}$, чтобы оценить Q-функцию от $(s_t, y_{t+1})$. 


![](pictures/9_1.png)

В эксперименте авторы хотели показать, что модель **ORGAN** способна генерировать молекулы с желаемыми свойствами, при этом поддерживая разнообразие генерируемых молекул, и сравнить результаты с другими моделями. Поэтому функция награды учитывает, смог ли дискриминатор отличить сгенерированную строку от настоящей $D(Y_{1:T})$, и свойства сгенерированной молекулы $O_i(Y_{1:T})$.


![](pictures/10.png)

Также вводится штраф за генерацию повторяющихся молекул.

Реализация архитектуры:

- Генератор: **RNN-сеть с ячейками LSTM**
- Дискриминатор: **сверточная нейронная сеть**.


**3. Датасет:**

В статье использовалась случайная выборка размером 5 000 молекул (каждая молекула может иметь до 9 тяжелых атомов (CONF))
из 134 000 малых устойчивых молекул [Ramakrishnan et al., 2014](https://www.nature.com/articles/sdata201422.pdf), которая в свою очередь является подвыборкой из [GDB-17 universe](https://pubs.acs.org/doi/pdf/10.1021/ci300415d),
состоящей из около 166 000 000 000 органических молекул.

**Кол-во молекул:** 5 000

**Процедура предобработки:**

- построили словарь, отображающий алфавит символов в числа. Размер алфавита: 43 символа.
- каждая smiles-строка с помощью словаря была переведена в вектор длины 51 (строки меньшей длины дополнялись символами “_” до необходимой длины)



**4. Измерение качества модели**


Сгенерированные молекулы оценивались по каждому из желаемых свойств, а также по свойству разнообразия по шкале от 0 до 1.

Желаемые свойства:
- Solubility (LogP) - растворимость. Посчитана с помощью RDKit’s Crippen function.
- Синтезируемость - насколько трудно синтезировать молекулу.
- Druglikeness - вероятность оказаться лекарством. Данное свойство коррелирует с предыдущими двумя.

Оценка разнообразия:
- Оценивалось сходство сгенерированной молекулы с тренировочным датасетом с помощью Jaccard distance для фингерпринтов молекул.
